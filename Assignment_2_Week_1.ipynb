```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs

df = pd.read_csv('../class.csv')
df.head()
```

```python
sns.distplot(df['texts_yesterday'].dropna(), kde=False) #Selecting the rating column. I need to drop the NA's for the plot
plt.title('Rating of Google Play apps')
plt.xlabel('Rating (stars)')
plt.show()
```

```python
ax = sns.scatterplot(x="instagram_followers", y="linkedin_con", hue="gender", data=df)
```

```python
g = sns.jointplot(x="phones_owned", y="apps_phone", data=df, kind="kde")
```

```python
sns.catplot(x="phones_owned", y="texts_yesterday", hue="os_phone", kind="box", data=df);
```

```python
sns.catplot(x="like_social", kind="count", palette="ch:.25", data=df);
```
