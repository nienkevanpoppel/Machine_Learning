```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs

df = pd.read_csv('../funda-amsterdam.csv')
df.head()
```

```python
df_subset= df[['price','bedrooms','year_built']]
print(df_subset)
```

```python
sns.pairplot(df_subset)
plt.show()
```

```python

df_subset2= df_subset[(df_subset["price"]>20000) & (df_subset["bedrooms"] < 10)]
print(df_subset2)
```

```python
sns.scatterplot(x="bedrooms", y="price", data=df_subset2)
plt.show()
```

Despite the values overlapping on the scatterplot, there can be a upward line drawn accross the values. This would mean a positive lineairity across both variables price and bedrooms. This means: the more bedrooms there are, the higher the price is.

```python
from scipy.stats.stats import pearsonr
print(pearsonr(df_subset2['price'],df_subset2['bedrooms']))
```

The output gives us two numbers: r which is 0.468, and p which appears to be 0, but which is actually just really small. We can conclude the following:
- Price and bedrooms are moderately correlated (r = 0.468)
- The correlation is highly significant (p < 0.001)
